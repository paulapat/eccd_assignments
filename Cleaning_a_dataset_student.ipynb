```python
def unify_victoria_secret(df):
    """
    We want that all brands that are related to Victoria's Secret
    have `victoria's secret` as their brand instead of what they
    currently have.
    """
    df = df.copy()
    new_string = "victoria's secret"
    # Write your code here
    # 1. Veo cuantos valores únicos hay de la variable nombre de marca
    df['brand_name'].unique()
    # 2. Creo una lista con los nombres de las marcas relacionadas con Victoria's Secret
    related_vs=["Victoria's Secret","Victoria's Secret Pink", "Victorias-Secret" ]
    # 3. Ahora unifico en un solo nombre las marcas vinculadas a VS
    df["brand_name"]= df["brand_name"].replace(related_vs, new_string)
    return df
```

## Cleaning up the price

In this sectino we are going to transform the `price` column into a float column in USD dolars.

For this, be careful of the different formats in the data.

For simplicity, you might assume that all the prices are in USD dolars, regarding of the symbol of the currency used.

```python
import re
```

```python
pd.set_option('display.max_rows', None)
p=df_unified['price'].unique()
for value in p:
    print(value)
```

```python
def clean_price(df):
    """
    In this function we will transform the
    `price` column into a column of floats.
    In case a product has more than one price,
    return the lowest one.
    """

    # Write your code here

    # 1. Me aseguro de considerar toda la información de la columna precio como string así las operaciones tipicas con string funcionan bien
    # 2. Saco espacios/saltos de línea que distorcionan el procesamiento. Trabajo de forma manual aquellas celdas que tienen un "Now" y me quedo con el precio de ahora que es el inferior
    # 3. Si en la columna de precio detecto un "-", me quedo con la primera parte que esta antes del "-", que es naturalmente el precio más bajo
    # 4. Tal como vimos en la clase del 4/11/2023, puedo usar expresiones regulares para quedarme con la parte númerica del texto (ignoro los simbolos de la moneda, etc.). Convierto a float.
    # 5. Aplico esta función en la columna de precio

    df['price'] = df['price'].astype(str)
    df['price'] = df['price'].str.strip()
    df['price'] = df['price'].apply(lambda x: ''.join(re.split(r'\s+', x.strip())))
    df['price'] = df['price'].str.strip()

    df['price'] = df['price'].str.replace(r'\$35\.00Now\$10\.00', "10.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$35\.00Now\$20\.00', "20.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$15\.00Now\$8\.00', "8.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$38\.00Now\$20\.00', "20.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$20\.00Now\$10\.00', "10.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$30\.00Now\$20\.00', "20.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$35\.00Now\$24\.00', "24.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$38\.00Now\$24\.00', "24.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$20\.00Now\$12\.00', "12.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$38\.00Now\$12\.00', "12.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$20\.00Now\$6\.00', "6.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$12\.00Now\$4\.00', "4.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$10\.00Now\$4\.00', "4.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$20\.00Now\$15\.00', "15.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$26\.00Now\$8\.00', "8.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$45\.00Now\$12\.00', "12.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$46\.00Now\$15\.00', "15.00", regex=True)
    df['price'] = df['price'].str.replace(r'\$30\.00Now\$8\.00', "8.00", regex=True)


    if "-" in df['price']:
        df['price'] = df['price'].str.split("-").str[0]

    def detecta_num(colp):

      extrae_num = re.search(r'\d+(\.\d+)?', colp)
      if extrae_num:
        return extrae_num.group(0)
      else:
          return None

    # Aplicamos la función detecta_num a la columna 'price'
    df['price'] = df['price'].apply(detecta_num)
    df['price'] = df['price'].astype(float)


    return df
```

```python
def low_high_product_mean(df):
    """
    Finally, we will calculate `product_category` with the lowest and highest mean price
    for the brand Victoria's Secret.
    """

    # Write your code here
    # 1. Me quedo con la marca VS (filtro la base)
    vs_df=df[df["brand_name"]=="victoria's secret"]
    mediaxcat=vs_df.groupby('product_category')['price'].mean()
    highest_mean=mediaxcat.max()
    lowest_mean=mediaxcat.min()
    return lowest_mean, highest_mean

```
